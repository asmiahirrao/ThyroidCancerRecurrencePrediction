```python
# Imports
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Update with your actual file path
data_path = r'C:/Users/aahir/OneDrive/Desktop/ThyroidCancerRecurrencePrediction/data/thyroid_cancer_data.csv'
df = pd.read_csv(data_path)

# Data Visualizations
sns.countplot(data=df, x='Recurred')
plt.show()

# Convert categorical columns to numeric (one-hot encoding) for correlation analysis
df_numeric = pd.get_dummies(df, drop_first=True)

# Correlation Analysis
plt.figure(figsize=(12, 10))
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm')
plt.show()
```
